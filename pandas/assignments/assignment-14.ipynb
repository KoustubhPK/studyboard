### __25. Delete rows, columns, and indices from a DataFrame using appropriate Pandas methods.__

```python
df = pd.DataFrame(np.random.randint(1,30, [6,5]), index=list("abcdef"), columns=list("ABCDF"))
df
```

- __Deleting Rows:__

```python
df.drop(["a","c"])
```

- __Deleting Column:__

```python
df.drop(["A", "C"], axis=1)
```

- __Deleting Indices:__

```python
df.reset_index(drop=True)
```

### __26. Rename DataFrame columns and indices using Pandas methods.__

- __Renaming Columns:__

```python
df.rename({"A": "Test-A", "C": "Test-C"}, axis=1, inplace=True)
df
```

- __Renaming Indices:__

```python
df.rename({"a": "test-a", "c": "test-c"}, axis=0)
```

### __27. Find and replace specific characters in DataFrame column names.__

```python
# Columns before changing specific character
df.columns
```

```python
# Logic for changing Specific character
names = []
for x in df.columns:
    individual = ""
    for i in x:
        if i == "T":
            individual += "B"
        else:
            individual += i
    names.append(individual)
df.columns = names
```

```python
# Specific character is replaced in Column names
df
```
