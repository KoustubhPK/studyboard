```python
import re
import warnings
import numpy as np
import pandas as pd
warnings.filterwarnings('ignore')
```

#### __Creating different dataframes__

```python
df1 = pd.DataFrame(np.random.randint(1,15, size=(6,5)), columns=list("ABCDE"))
df1
```

```python
df2 = pd.DataFrame(np.random.randint(16,30, size=(6, 4)), columns=list("ABCD"))
df2
```

```python
df3 = pd.DataFrame(np.random.randint(31,45, size=(6,5)), columns=list("ABCDE"))
df3
```

#### __Concat__

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], ignore_index=True)
```

```python
pd.concat([df1, df2, df3], axis=1)
```

#### __Joins__

```python
pd.concat([df1,df2],join = 'outer')
```

```python
pd.concat([df1,df2],join = 'inner')
```

```python
df3.columns = list('ACXYZ')
df3
```

```python
pd.concat([df2, df3])
```

```python
pd.concat([df2, df3], join="inner")
```

#### __Joins__

```python
df2, df3
```

```python
pd.merge(df2, df3, how="left")
```

```python
df2.merge(df3, how="outer")
```
