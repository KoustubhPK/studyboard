## __6. Access items by using for loop:__

### __6.1. df.items()__

```python
for col, val in df.items():
    print(f"Column Name: {col}\n{'-'*40}\n{val}\n{'-'*40}")
```

### __6.2. df.iterrows()__

```python
for row, val in df.iterrows():
    print(f"Row Name: {row}\n{'-'*40}\n{val}\n{'-'*40}")
```
