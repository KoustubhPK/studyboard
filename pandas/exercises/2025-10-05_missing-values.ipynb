# __Missing Values__

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### __Reading Data from CSV__

```python
emp_data = pd.read_csv("data/Emp_data.csv")

# Adding NaN valuse to column - 3
for x in emp_data.iloc[:5:2, 3]:
    emp_data.iloc[:5:2, 3] = np.nan

# Making the float value to String
emp_data['Weight in Kgs'][1] = '60'

emp_data.head(3)
```

### __Changed dtype Float to Object__

```python
emp_data['Weight in Kgs'].info()
```

### __Replacing _NaN_ to 0__

```python
emp_data['Weight in Kgs'] = emp_data['Weight in Kgs'].replace(float('NaN'), 0)
emp_data.head(5)
```

### __Converting all Weight values to int__

```python
emp_data['Weight in Kgs'] = emp_data['Weight in Kgs'].apply(lambda x: int(x))
```

### __Missing Weight Values Replacing by mean()__

```python
emp_data['Weight in Kgs'].replace(0, emp_data['Weight in Kgs'].mean(), inplace=True)
emp_data.head(5)
```

### __Converting all Weight values to int__

```python
emp_data['Weight in Kgs'] = emp_data['Weight in Kgs'].astype(int)
```
